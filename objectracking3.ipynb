```python
. BOOSTING Tracker
Açıklama: Bu algoritma, AdaBoost algoritmasına dayanan bir izleyicidir. Boosting algoritması, nesne tanıma görevlerinde
yaygın olarak kullanılır.
Avantajları:
Zayıf sınıflandırıcıları birleştirerek güçlü bir izleyici oluşturur.
Nesne tanıma üzerine temellendirilmiştir.
Dezavantajları:
Yüksek hesaplama maliyetine sahiptir.
Diğer algoritmalara göre daha düşük doğruluğa sahip olabilir.
Kullanım Durumu: Yüksek doğruluk ve nesne tanıma temelli takip uygulamalarında kullanılabilir.
    
    
2. MIL (Multiple Instance Learning) Tracker
Açıklama: MIL izleyici, izleme sırasında yanlış pozitif durumlarla başa çıkmak için birden fazla örnek öğrenme yöntemini 
kullanır. Yani, izleme algoritması yalnızca pozitif bir örneği değil, aynı zamanda o bölgedeki birden fazla örneği öğrenir.
Avantajları:
Daha sağlamdır ve öğrenme aşamasında birden fazla örneği göz önüne alarak hatalı izlemeyi azaltır.
Dezavantajları:
Diğer algoritmalara göre daha yavaştır.
Kullanım Durumu: Karışık sahnelerde veya nesnenin görünümünde büyük değişiklikler olduğunda kullanılabilir
    
.
3. KCF (Kernelized Correlation Filters) Tracker
Açıklama: KCF izleyici, çekirdekli korelasyon filtreleri kullanarak nesne izleme sağlar. Bu algoritma, nesnenin hızla
    izlenmesine olanak tanır ve Fourier dönüşümleri kullanarak verimli bir şekilde çalışır.
Avantajları:
Oldukça hızlıdır.
Gerçek zamanlı uygulamalar için uygundur.
Dezavantajları:
Ölçek değişikliklerini iyi takip edemez (nesne büyürse veya küçülürse izleyici sorun yaşayabilir).
Kullanım Durumu: Nesnenin boyutunun sabit olduğu ve hızlı izleme gerektiren senaryolarda tercih edilir.
    

4. TLD (Tracking, Learning, and Detection) Tracker
Açıklama: TLD, izleme sırasında nesneyi öğrenen ve aynı zamanda nesneyi yeniden tespit etmeye çalışan bir izleyicidir. 
Bu, izleyicinin kaybolan nesneleri tekrar bulabilme yeteneği sağlar.
Avantajları:
Öğrenme bileşeni sayesinde nesne kaybolsa bile yeniden tespit edebilir.
Nesne kaybolup tekrar görünse bile izleme devam edebilir.
Dezavantajları:
Diğer izleyicilere göre daha yavaştır.
Kullanım Durumu: Nesnenin kaybolup tekrar ortaya çıkabileceği senaryolarda kullanılır.
5. MedianFlow Tracker
Açıklama: MedianFlow, nesnenin piksel hareketini izleyerek medyan akışını kullanır. Bu algoritma, kısa vadeli hareketleri çok hassas bir şekilde izler ve izleme sırasında geri adım atabilir.
Avantajları:
Kısa süreli izleme için oldukça hassastır.
Küçük nesnelerde iyi sonuçlar verir.
Dezavantajları:
Nesnenin hızlı hareket ettiği senaryolarda başarısız olabilir.
Kullanım Durumu: Yavaş hareket eden veya sabit bir nesnenin kısa süreli izlenmesinde kullanılır.
6. GOTURN Tracker
Açıklama: GOTURN, derin öğrenmeye dayalı bir izleme algoritmasıdır. Pretrained bir konvolüsyonel sinir ağı (CNN) kullanarak nesneyi izler. Bu yöntem, nesnenin görünümünün çok değiştiği durumlarda oldukça etkilidir.
Avantajları:
Nesnenin görünümünün çok fazla değiştiği durumlarda iyi çalışır.
Derin öğrenme tabanlıdır ve karmaşık sahnelerde izleme yapabilir.
Dezavantajları:
Diğer izleyicilere göre çok daha fazla hesaplama gücü gerektirir.
Kullanım Durumu: Görünüm değişikliklerinin sık olduğu ve derin öğrenme tabanlı izleme gereken senaryolarda kullanılır.
7. CSRT (Discriminative Correlation Filter with Channel and Spatial Reliability) Tracker
Açıklama: CSRT, KCF izleyiciye göre geliştirilmiş bir sürümdür. Uzaysal güvenilirlik haritaları kullanarak nesnenin ölçek ve konumunu daha doğru bir şekilde takip eder.
Avantajları:
Nesnenin boyut ve konum değişikliklerine daha duyarlıdır.
KCF izleyiciye göre daha hassas ve güvenilirdir.
Dezavantajları:
Hızı KCF'ye göre biraz daha düşüktür, ancak yine de gerçek zamanlı izleme için uygundur.
Kullanım Durumu: Hem hız hem de doğruluk gereken senaryolarda, özellikle nesnenin boyutunun değiştiği durumlarda kullanılabilir.
8. MOSSE (Minimum Output Sum of Squared Error) Tracker
Açıklama: MOSSE izleyici, oldukça hızlı ve düşük hesaplama maliyeti olan bir izleyicidir. Hedef nesneye dair minimum hata ile en uygun filtreyi bulmaya çalışır.
Avantajları:
Çok hızlıdır ve düşük kaynak gereksinimine sahiptir.
Gerçek zamanlı izleme uygulamaları için uygundur.
Dezavantajları:
Diğer izleyicilere göre daha az doğru olabilir.
Kullanım Durumu: Hızlı izleme gereken ancak yüksek doğruluk beklentisinin olmadığı durumlarda tercih edilebilir.
    
    
Sonuç
Bu izleme yöntemlerinin her biri, farklı koşullar ve uygulamalar için uygundur. İşlem hızına mı, doğruluğa mı yoksa nesnenin değişen görünümüne mi öncelik verilmesi gerektiğine bağlı olarak doğru izleme algoritmasını seçmek önemlidir:

Hız ve düşük kaynak kullanımı istiyorsanız, MOSSE veya KCF izleyici ideal olabilir.
Doğruluk ve nesnenin kaybolduğunda yeniden bulunması önemliyse, CSRT veya TLD gibi algoritmaları tercih edebilirsiniz.
Derin öğrenme kullanmak ve görünüm değişikliklerini izlemek için GOTURN uygun olacaktır.
Her algoritma farklı avantajlar ve dezavantajlar sunar, bu yüzden spesifik bir uygulamaya en uygun olanını seçmek performansı
önemli ölçüde artırabilir.






```

```python
import numpy
import cv2
```

```python
def ask_for_tracker():
    print("Welcome! What Tracker API would you like to use?")
    print("Enter 0 for BOOSTING: ")
    print("Enter 1 for MIL: ")
    print("Enter 2 for KCF: ")
    print("Enter 3 for TLD: ")
    print("Enter 4 for MEDIANFLOW: ")
    choice = input("Please select your tracker: ")
    
    if choice == '0':
        tracker = cv2.TrackerBoosting_create()
    if choice == '1':
        tracker = cv2.TrackerMIL_create()
    if choice == '2':
        tracker = cv2.TrackerKCF_create()
    if choice == '3':
        tracker = cv2.TrackerTLD_create()
    if choice == '4':
        tracker = cv2.TrackerMedianFlow_create()


    return tracker
```

```python
tracker = ask_for_tracker()
tracker_name = str(tracker).split()[0][1:]

# Read video
cap = cv2.VideoCapture(0)

# Read first frame.
ret, frame = cap.read()


# Special function allows us to draw on the very first frame our desired ROI
roi = cv2.selectROI(frame, False)

# Initialize tracker with first frame and bounding box
ret = tracker.init(frame, roi)

while True:
    # Read a new frame
    ret, frame = cap.read()
    
    
    # Update tracker
    success, roi = tracker.update(frame)
    
    # roi variable is a tuple of 4 floats
    # We need each value and we need them as integers
    (x,y,w,h) = tuple(map(int,roi))
    
    # Draw Rectangle as Tracker moves
    if success:
        # Tracking success
        p1 = (x, y)
        p2 = (x+w, y+h)
        cv2.rectangle(frame, p1, p2, (0,255,0), 3)
    else :
        # Tracking failure
        cv2.putText(frame, "Failure to Detect Tracking!!", (100,200), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,0,255),3)

    # Display tracker type on frame
    cv2.putText(frame, tracker_name, (20,400), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0),3);

    # Display result
    cv2.imshow(tracker_name, frame)

    # Exit if ESC pressed
    k = cv2.waitKey(1) & 0xff
    if k == 27 : 
        break
        
cap.release()
```

```python
roi = cv2.selectROI(frame, False)
Bu satır, kullanıcının video karesi üzerinde elle bir ROI (Region of Interest) seçmesini sağlar.
selectROI fonksiyonu, bir pencere açarak kullanıcının fare ile bir dikdörtgen çizmesine olanak tanır. 
Seçilen alan (ROI), izlenecek nesne olur.
False parametresi, bu pencerenin yeniden boyutlandırılabilir olmadığını belirtir.

ret = tracker.init(frame, roi)
Burada, izleyiciye (tracker) ilk video karesi ve kullanıcının seçtiği ROI (dikdörtgen) verilir. İzleme işlemi başlar.
tracker.init() fonksiyonu, izleyiciyi başlatır ve seçilen nesneyi izlemeye başlar

success, roi = tracker.update(frame)
tracker.update(frame) fonksiyonu, izleyicinin yeni video karesi üzerinde izleme yapmasını sağlar.
success değişkeni, izleyicinin nesneyi başarılı bir şekilde takip edip etmediğini belirtir 
(eğer True ise izleme başarılı olmuştur).
roi değişkeni, nesnenin güncellenmiş (takip edilen) koordinatlarını (x, y, w, h) içerir.
```
